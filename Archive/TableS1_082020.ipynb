```r
summary_stats_all_subjects = function(conc_list){
    #"""
    #:param: takes list of concentration values
    #:output: a list of descriptive stats (median, min, max, sd)

    #"""
    stats_list = c(mean(conc_list), median(conc_list),min(conc_list),max(conc_list),sd(conc_list))
    return(stats_list)
}
```

```r
#calling summary_stats_all_subjects function from above
Eotaxin = summary_stats_all_subjects(Eotaxin_list)
Eotaxin3 = summary_stats_all_subjects(Eotaxin3_list)
Fractalkine = summary_stats_all_subjects(Fractalkine_list)
I309 = summary_stats_all_subjects(I309_list)
IFNg = summary_stats_all_subjects(IFNg_list)
IL10 = summary_stats_all_subjects(IL10_list)
IL12p70 = summary_stats_all_subjects(IL12p70_list)
IL13 = summary_stats_all_subjects(IL13_list)
IL17B = summary_stats_all_subjects(IL17B_list)
IL17D = summary_stats_all_subjects(IL17D_list)
IL1b = summary_stats_all_subjects(IL1b_list)
IL4 = summary_stats_all_subjects(IL4_list)
IL6 = summary_stats_all_subjects(IL6_list)
IL8 = summary_stats_all_subjects(IL8_list)
IP10 = summary_stats_all_subjects(IP10_list)
ITAC = summary_stats_all_subjects(ITAC_list)
MCP1 = summary_stats_all_subjects(MCP1_list)
MIG = summary_stats_all_subjects(MIG_list)
MIP1a = summary_stats_all_subjects(MIP1a_list)
MIP1b = summary_stats_all_subjects(MIP1b_list)
TARC = summary_stats_all_subjects(TARC_list)
TNFa = summary_stats_all_subjects(TNFa_list)  
```

```r
table_summary = function(){
    #"""
    #Putting stats into a table(matrix) to make transfer to excel easier.

    #:param: none
    #:output: a 22x4 matrix containing all descriptive statistics for each cytokine. 
    #"""
    compartment_stats_array_initial = cbind(Eotaxin, Eotaxin3, Fractalkine, I309,
                                           IFNg, IL10, IL12p70, IL13,
                                           IL17B, IL17D, IL1b, IL4, IL6,
                                           IL8, IP10, ITAC, MCP1, MIG,
                                           MIP1a, MIP1b, TARC, TNFa)
    compartment_stats_array = t(compartment_stats_array_initial)
    
    colnames(compartment_stats_array) = c('Mean','Median', 'Min', 'Max', 'STD')
    return(compartment_stats_array)
}
```
