```r
get_individual_sums = function(table){
    #"""
    # Calculating the sum of the proprotions for each individual.

    #:param: table (dataframe of proportions)
    #:output: a 28x22 matrix containing all descriptive statistics for each cytokine. 
    #"""
    for (i in 1:length(table$I309)){
        if(table$I309[i] == 'NaN'){
            table$I309[i] = 0 #setting NaNs as zeros to get sum
        }
    }
    individual_sum = rowSums(table)
    table_with_sums = data.frame(cbind(table,individual_sum))
}

NLF_proportions_df = get_individual_sums(individual_proportions_NLF)
ELF_proportions_df = get_individual_sums(individual_proportions_ELF)
Sputum_proportions_df = get_individual_sums(individual_proportions_Sputum)
Serum_proportions_df = get_individual_sums(individual_proportions_Serum)
```
